```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("new_train.csv")
df.info()
```

# Grafik Batang

```python
x = list(df.iloc[:,1])
y = list(df.iloc[:,1])
```

```python
sns.barplot(
    data = df,
    x    = 'marital', 
    y    = 'age',
)
plt.xlabel('Status')
plt.title('Grafik Untuk Menunjukan Status Seseorang Berdasarkan Umur')
plt.show()
```

### bisa kita lihat bahwa kebanyakan orang pada umur dibawah 30 tahun itu masih berstatus single

# Pie Chart

```python
df = df.dropna(subset=['job'])
data_pie = df['job'].value_counts().rename_axis('job').reset_index(name='age')
```

```python
# plt.figure(figsize=(15,20))
# data_pie['job'].plot.pie(autopct='%.2f%%')
plt.figure(figsize=(10,10))
plt.pie(
    data_pie.age, 
    labels     = data_pie.job, 
    startangle = 90,
    autopct    = '%.1f%%'
)
plt.title('Data Orang Yang Memiliki Pekerjaan')
plt.show()
```

### dari chart di atas bisa kita simpulkan bahwa pekerjaan admin mempunyai populasi yang paling besar

# Scatter Plot

```python
sns.scatterplot(
    data = df,
    x='education',
    y='age',
    hue = 'job'

)
```

### dari grafik di atas dapat kita ketahui bahwa pekerjaan entrepeneur mempunyai umur yang paling tua

# Heatmap

```python
atribut_numerik = ['age', 'duration','campaign']
```

```python
correlation_matrix = df[atribut_numerik].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Atribut Numerik dan diabetes')
plt.show()
```

### bisa kita lihat nilai yang paling tinggi di kolom ,age,duration,dan campaign adalah 1.00

# Box Plot

```python
plt.figure(figsize=(10,20))
sns.boxplot(
    data   = df,
    x      = 'job',
    y      = 'age',
)

plt.xlabel('pekerjaan')
plt.ylabel('age')
plt.xticks(rotation = 45 , ha = 'right')
plt.title('Orang yang Memiliki Pekerjaan berdasarkan umur')
plt.show()
```

### bisa kita lihat pada grafik di atas bahwa pekerjaan retired/pensiun yang paling tua
